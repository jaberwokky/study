```python
import numpy as np
```

```python
grouped = data.groupby(['cgroup', 'created-ym'])

cohorts = grouped.agg({'client_id': pd.Series.nunique,
                       'id': pd.Series.nunique,
                       'final_price': np.sum})

cohorts.rename(columns={'client_id': 'totalclients',
                        'id': 'totalorder'}, inplace=True)
cohorts.head()
```

```python
cohorts
```

```python
def cohort_period(df):
    df['cperiod'] = np.arange(len(df)) + 1
    return df

cohorts = cohorts.groupby(level=0).apply(cohort_period)
cohorts.head()
```

```python
cohorts.reset_index(inplace=True)
cohorts.set_index(['cgroup', 'cperiod'], inplace=True)

cohort_group_size = cohorts['totalclients'].groupby(level=0).first()
cohort_group_size.head()
```

```python
cohorts['totalclients'].head()
```

```python
cohorts['totalclients'].unstack(0).head()
```

```python
user_retention = cohorts['totalclients'].unstack(0).divide(cohort_group_size, axis=1)
user_retention.head(10)
```

```python
user_retention[['2018-10', '2018-11', '2018-12']].plot(figsize=(10,5))
plt.title('Cohorts: User Retention')
plt.xticks(np.arange(1, 12.1, 1))
plt.xlim(1, 12)
plt.ylabel('% of Cohort Purchasing');
```

```python
import seaborn as sns
sns.set(style='white')

plt.figure(figsize=(24, 16))
plt.title('Cohorts: User Retention')
sns.heatmap(user_retention.T, mask=user_retention.T.isnull(), annot=True, fmt='.0%');
```

```python
import seaborn as sns
sns.set(style='ticks')
plt.figure(figsize=(24, 16))
plt.title('Cohorts: Orders Per User')
sns.heatmap(user_retention.T, mask=user_retention.T.isnull(), annot=True, fmt='.3');
```

```python
def jsPoint(lat, long):
    import math
    if (math.isnan(lat) or math.isnan(long)):
        return Point([])
    else:
        return Point((lat, long))
```
